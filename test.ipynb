```python
import numpy as np
import pandas as pd
import json
import os
from sklearn.metrics.pairwise import cosine_similarity
from sentence_transformers import SentenceTransformer
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
# Step 1: Convert answers to embeddings
def convert_to_embeddings(data):
    print(f"Converting data to embeddings: {data}")
    # For simplicity, we'll convert each string into a basic embedding using ASCII sum
    # In a real case, this would use a model like Word2Vec, BERT, etc.
    # embeddings = [np.array([sum(ord(c) for c in answer)]) for answer in data]
    # print(f"Embeddings: {embeddings}")
    # return embeddings


    print("Starting embedding conversion...")
    embeddings = model.encode(data)
    print("Finished embedding conversion.")
    return embeddings

# Step 2: Calculate cosine similarity
def calculate_cosine_similarity(embedding1, embedding2):
    # Reshape the embeddings for cosine similarity calculation
    print(f"Calculating cosine similarity between {embedding1} and {embedding2}")
    embedding1 = np.reshape(embedding1, (1, -1))
    embedding2 = np.reshape(embedding2, (1, -1))
    
    similarity = cosine_similarity(embedding1, embedding2)[0][0]
    # Adjust negative cosine similarity
    similarity = max(similarity, 0)
    print(f"Cosine similarity: {similarity}")
    return similarity

# Step 3: Scoring based on similarity
def score_from_similarity(similarity, total_marks):
    print(f"Calculating score from similarity {similarity} with total marks {total_marks}")
    score = similarity * total_marks
    print(f"Score: {score}")
    return score

# Step 4: Calculate total marks
def calculate_total_marks(scores):
    print(f"Calculating total marks from scores: {scores}")
    total_marks = sum(scores)
    print(f"Total marks: {total_marks}")
    return total_marks

# Helper function to save intermediate results
def save_to_file(data, filename, folder="results"):
    if not os.path.exists(folder):
        os.makedirs(folder)
    path = os.path.join(folder, filename)
    print(f"Saving data to {path}")
    with open(path, 'w') as f:
        f.write(str(data))

# Step 5: Process each student
def process_student(student, answer_key, save_intermediate=False):
    print(f"Processing student: {student['student_id']}")
    scores = []
    
    # Iterate over each set of answers (2-mark, 16-mark, 8-mark)
    for idx, student_answers in enumerate(student['answers']):
        part_marks = []
        for q_idx, student_answer in enumerate(student_answers):
            # Convert both student and answer key to embeddings
            student_emb = convert_to_embeddings([student_answer])[0]
            answer_emb = convert_to_embeddings([answer_key[idx][q_idx]])[0]
            
            # Calculate cosine similarity
            similarity = calculate_cosine_similarity(student_emb, answer_emb)
            
            # Calculate score
            total_marks = 2 if idx == 0 else (16 if idx == 1 else 8)
            score = score_from_similarity(similarity, total_marks)
            part_marks.append(score)
            
            # Save intermediate steps if required
            if save_intermediate:
                save_to_file({"student_id": student['student_id'], "question": q_idx + 1, "part": idx + 1, "score": score}, f"{student['student_id']}_q{q_idx + 1}_score.txt")
        
        scores.append(part_marks)
    
    # Calculate total scores for each part
    part_a_marks = calculate_total_marks(scores[0])
    part_b_marks = calculate_total_marks(scores[1])
    part_c_marks = calculate_total_marks(scores[2])
    total_marks = part_a_marks + part_b_marks + part_c_marks

    if save_intermediate:
        save_to_file({"student_id": student['student_id'], "part_a": part_a_marks, "part_b": part_b_marks, "part_c": part_c_marks, "total": total_marks}, f"{student['student_id']}_final_marks.txt")
    
    print(f"Total marks for student {student['student_id']}: {total_marks}")
    return {
        "student_id": student['student_id'],
        "part_a_marks": part_a_marks,
        "part_b_marks": part_b_marks,
        "part_c_marks": part_c_marks,
        "total_marks": total_marks
    }

# Step 6: Calculate all students' results
def calculate_results(students, answer_key, save_intermediate=False):
    print("Calculating results for all students...")
    results = []
    
    for student in students:
        result = process_student(student, answer_key, save_intermediate)
        results.append(result)
    
    # Convert results to a DataFrame
    df = pd.DataFrame(results)
    
    if save_intermediate:
        df.to_csv("final_results.csv", index=False)
        print(f"Final results saved to final_results.csv")
    
    print(df)
    return df

# Call the main function
calculate_results(students, answer_key, save_intermediate=True)
```

```python
# Step 1: Convert answers to embeddings
def convert_to_embeddings(data):
    print(f"Converting data to embeddings: {data}")
    # Assuming `model` is a pre-trained model you are using for embeddings
    print("Starting embedding conversion...")
    embeddings = model.encode(data)
    print("Finished embedding conversion.")
    return embeddings

# Step 2: Calculate cosine similarity
def calculate_cosine_similarity(embedding1, embedding2):
    print(f"Calculating cosine similarity between {embedding1} and {embedding2}")
    embedding1 = np.reshape(embedding1, (1, -1))
    embedding2 = np.reshape(embedding2, (1, -1))
    
    similarity = cosine_similarity(embedding1, embedding2)[0][0]
    # Adjust negative cosine similarity
    similarity = max(similarity, 0)
    print(f"Cosine similarity: {similarity}")
    return similarity

# Step 3: Scoring based on similarity
def score_from_similarity(similarity, total_marks):
    print(f"Calculating score from similarity {similarity} with total marks {total_marks}")
    score = similarity * total_marks
    print(f"Score: {score}")
    return score

# Step 4: Calculate total marks
def calculate_total_marks(scores):
    print(f"Calculating total marks from scores: {scores}")
    total_marks = sum(scores)
    print(f"Total marks: {total_marks}")
    return total_marks

# Helper function to save intermediate results
def save_to_file(data, filename, folder="results"):
    if not os.path.exists(folder):
        os.makedirs(folder)
    path = os.path.join(folder, filename)
    print(f"Saving data to {path}")
    with open(path, 'w') as f:
        f.write(str(data))

# Step 5: Process each student
def process_student(student, answer_key, save_intermediate=False):
    print(f"Processing student: {student['student_id']}")
    scores = []
    question_scores = []

    # Iterate over each set of answers (2-mark, 16-mark, 8-mark)
    for idx, student_answers in enumerate(student['answers']):
        part_marks = []
        part_scores = []  # To hold individual question scores for this part
        for q_idx, student_answer in enumerate(student_answers):
            # Convert both student and answer key to embeddings
            student_emb = convert_to_embeddings([student_answer])[0]
            answer_emb = convert_to_embeddings([answer_key[idx][q_idx]])[0]
            
            # Calculate cosine similarity
            similarity = calculate_cosine_similarity(student_emb, answer_emb)
            
            # Calculate score
            total_marks = 2 if idx == 0 else (16 if idx == 1 else 8)
            score = score_from_similarity(similarity, total_marks)
            part_marks.append(score)
            part_scores.append({"question": q_idx + 1, "score": score})  # Store question score
            
            # Save intermediate steps if required
            if save_intermediate:
                save_to_file({"student_id": student['student_id'], "question": q_idx + 1, "part": idx + 1, "score": score}, f"{student['student_id']}_q{q_idx + 1}_score.txt")
        
        scores.append(part_marks)
        question_scores.append(part_scores)  # Append scores for this part
    
    # Calculate total scores for each part
    part_a_marks = calculate_total_marks(scores[0])
    part_b_marks = calculate_total_marks(scores[1])
    part_c_marks = calculate_total_marks(scores[2])
    total_marks = part_a_marks + part_b_marks + part_c_marks

    if save_intermediate:
        save_to_file({"student_id": student['student_id'], "part_a": part_a_marks, "part_b": part_b_marks, "part_c": part_c_marks, "total": total_marks}, f"{student['student_id']}_final_marks.txt")
    
    print(f"Total marks for student {student['student_id']}: {total_marks}")
    return {
        "student_id": student['student_id'],
        "part_a_marks": part_a_marks,
        "part_b_marks": part_b_marks,
        "part_c_marks": part_c_marks,
        "total_marks": total_marks,
        "question_scores": question_scores  # Include question scores
    }
```

```python
# Step 6: Calculate all students' results
def calculate_results(students, answer_key, save_intermediate=False):
    print("Calculating results for all students...")
    results = []

    for student in students:
        result = process_student(student, answer_key, save_intermediate)
        results.append(result)

    # Convert results to a DataFrame
    df = pd.DataFrame(results)

    # Now, let's add question scores
    question_scores = []
    for idx in range(len(answer_key)):  # Assuming answer_key has parts
        part_question_scores = [result['question_scores'][idx] for result in results]
        question_scores.append(part_question_scores)

    # Check the shapes of the collected scores
    print(f"Number of students: {len(students)}")
    print(f"Collected question scores structure (before DataFrame): {question_scores}")
    
    # Print lengths of individual parts
    for i, part in enumerate(question_scores):
        print(f"Part {i+1} has {len(part)} scores.")

    # Create a DataFrame for question scores
    question_scores_df = pd.DataFrame(question_scores).T  # Transpose to match student entries
    print(f"Shape of question scores DataFrame before assigning columns: {question_scores_df.shape}")

    # Create column names based on the answer_key structure
    total_questions = sum(len(part) for part in answer_key)
    print(f"Total expected columns based on answer key: {total_questions}")

    # Check the number of columns in the question scores DataFrame
    print(f"Number of columns in question_scores_df: {question_scores_df.shape[1]}")

    # Check if the lengths match before assigning columns
    if question_scores_df.shape[1] != total_questions:
        print("Warning: Length mismatch! Adjusting the number of question score columns.")
        question_scores_df = question_scores_df.iloc[:, :total_questions]  # Adjust if needed
        print(f"Adjusted shape of question_scores_df: {question_scores_df.shape}")

    # Create new column names for question scores
    new_column_names = [f'part_{i+1}_question_{q+1}_score' 
                        for i in range(len(answer_key)) 
                        for q in range(len(answer_key[i]))]
    
    print(f"New column names length: {len(new_column_names)}")
    print(f"Question scores DataFrame shape: {question_scores_df.shape}")

    # Ensure we do not assign more columns than available
    if len(new_column_names) != question_scores_df.shape[1]:
        print(f"Length mismatch! New column names: {len(new_column_names)}, Existing: {question_scores_df.shape[1]}")
        new_column_names = new_column_names[:question_scores_df.shape[1]]  # Adjust the length

    question_scores_df.columns = new_column_names

    # Combine both DataFrames
    df = pd.concat([df, question_scores_df], axis=1)

    if save_intermediate:
        df.to_csv("final_results.csv", index=False)
        print(f"Final results saved to final_results.csv")

    print(df)
    return df

# Call the main function
calculate_results(students, answer_key, save_intermediate=True)
```

```python
import ast

# Function to process the student answers and calculate the scores
def calculate_results(students, answer_key, save_intermediate=False):
    # Create a DataFrame from the students' data
    df = pd.DataFrame(students)
    
    # Initialize an empty list to store question scores
    question_scores_list = []
    
    # Process each student's scores and compute the required metrics
    for index, row in df.iterrows():
        # Convert the question scores from string representation to a Python list of dicts
        question_scores = ast.literal_eval(row['question_scores'])
        
        # Extract scores for each part and question
        part_a_scores = [q['score'] for q in question_scores[0]]  # Part A questions
        part_b_scores = [q['score'] for q in question_scores[1]]  # Part B questions
        part_c_scores = [q['score'] for q in question_scores[2]]  # Part C questions
        
        # Calculate the required scores
        total_part_a_score = sum(part_a_scores)
        total_part_b_score = sum(part_b_scores)
        total_part_c_score = sum(part_c_scores)
        
        # Store the scores in the list
        question_scores_list.append({
            'student_id': row['student_id'],
            'total_part_a_marks': total_part_a_score,
            'total_part_b_marks': total_part_b_score,
            'total_part_c_marks': total_part_c_score,
        })
    
    # Convert the question scores list to a DataFrame
    question_scores_df = pd.DataFrame(question_scores_list)
    
    # Merge the results with the original DataFrame
    df = pd.merge(df, question_scores_df, on='student_id')
    
    # Rename the columns to be more descriptive
    df.rename(columns={
        'total_part_a_marks': 'part_a_marks',
        'total_part_b_marks': 'part_b_marks',
        'total_part_c_marks': 'part_c_marks',
    }, inplace=True)
    
    # Add total marks column
    df['total_marks'] = df[['part_a_marks', 'part_b_marks', 'part_c_marks']].sum(axis=1)
    
    # Save intermediate results if needed
    if save_intermediate:
        df.to_csv('intermediate_results.csv', index=False)

    return df

results_df = calculate_results(students, answer_key=None, save_intermediate=False)
results_df
```
